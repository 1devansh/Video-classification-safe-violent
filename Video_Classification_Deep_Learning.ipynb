```

# compiling the model
model.compile(loss='categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])
```

```
# training the model
model.fit(X_train, train_label, epochs=10, validation_data=(X_val, val_label), callbacks=[mcp_save], batch_size=64)
```

# **Video prediction for any input video**

```
def print_class_from_prediction(prediction):
  
  dictt ={}
  dictt['Safe'] = prediction[0]
  dictt['Potentially suspicious'] = prediction[1]
  dictt['Violent'] = prediction[2]
  
  sorted_lps = sorted(
            dictt.items(),
            key=operator.itemgetter(1),
            reverse=True
        )
  
  result =[]
  for i, class_prediction in enumerate(sorted_lps):
            if i > 3 - 1 or class_prediction[1] == 0.0:
                break
            print("%s: %.2f" % (class_prediction[0], class_prediction[1]))
            result.append("%s: %.2f" % (class_prediction[0], class_prediction[1]))

  return result
```

```
capture = cv2.VideoCapture(os.path.join("/content/drive/My Drive/sem 6 deep Learning/new test/WhatsApp Video 2020-03-27 at 12.24.58 PM.mp4"))
width = capture.get(cv2.CAP_PROP_FRAME_WIDTH)   # float
height = capture.get(cv2.CAP_PROP_FRAME_HEIGHT) # float

fourcc = cv2.VideoWriter_fourcc(*'XVID')
video_writer = cv2.VideoWriter("result6.avi", fourcc, 15, (int(width), int(height)))

frames = []
frame_count = 0
while True:
    ret, frame = capture.read()
    # Bail out when the video file ends
    if not ret:
        break

    # Save each frame of the video to a list
    frame_count += 1
    frames.append(frame)

    if frame_count < 40:
        continue # capture frames untill you get the required number for sequence
    else:
        frame_count = 0

    # For each frame extract feature and prepare it for classification
    sequence = []
    for img in frames:
        image = cv2.resize(img,dsize=(300,300))
        image = np.array([img_to_array(image)
                      ]).astype('float32')
        image /= 255
        features = base_inception.predict(image)
        
        sequence.append(features)

    # Clasify sequence
    
    #prediction_images = prediction_images.reshape(prediction_images.shape[0], 8*8, 2048)
    feature = features.reshape(features.shape[0],8*8,2048)
    prediction = model.predict(feature)
    #prediction = saved_LSTM_model.predict(np.expand_dims(sequence, axis=0))
    #print(prediction)
    
    values = print_class_from_prediction(np.squeeze(prediction, axis=0))
    print(values)
    # Add prediction to frames and write them to new video
    for image in frames:
        for i in range(len(values)):
            cv2.putText(image, values[i], (40, 40 * i + 40), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), lineType=cv2.LINE_AA)
        video_writer.write(image)

    frames = []

video_writer.release()
```
